I checked the data type for the columns in the dataframe, I decided to change the data type of the user_id to object, because althoughthe user_id is a number, it's an identifier and we don't treat it as an integer. 

```python
# checkign data types:
customer_data.dtypes
```

```python
# checking for missing values:
customer_data.isnull().sum()
```

- Based on the results of the above query, the first name column has 11,259 missing values. This is interesting because there are no missing values in the last_name column. These missing values make up 5.45% of the total number of the rows. For this reason I decided to replace these missing values with "no_name". 

```python
# fill in the missing values in the first_name column with no_name:
customer_data['first_name'].fillna("no_name", inplace = True) 
```
