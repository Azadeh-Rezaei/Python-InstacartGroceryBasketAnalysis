# Task 4.7:

Question 2: Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

Answer: I used a for loop here and based on the results of the traffic on different days, I assigned the "Busiest day" label to Saturdays and Sundays (0 and 1) and the "Least Busy" to Tuesday and Wednesday (3 and 4) and the rest "Regular Busy". 
I then appended the list to my original dataframe.

```python
# Question 2: Created an empty list to save the results of the for-loop:
result_2 = []

for value in df_ords_prods_merged['order_day_of_week']:
    if value in (3, 4):
        result_2.append('Least Busy')
    elif value in (0, 1):
        result_2.append('Buesiest Day')
    else:
        result_2.append('Regular Busy')
```

```python
# Question 3: Checking the results of my for-loop in the list I created. 
result_2[:20]
```

```python
# Creating a new column to save the results of the for-loop from the result_2 list
df_ords_prods_merged['slowest/busiest days'] = result_2
```

Question 4: When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

Answer: I will start by looking at the frequeencis of the "order_hour_of_day". Then I will decide which hours to label as "Most Orders", "Average Orders", amd "Fewest Orders".

```python
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

Based on the above frequencies I am going to assign the labels as:

"Most Orders": 10 - 16. (hours with frequencies above 2,500,000)

"Average Orders": 7, 8, 9, 17, 18, 19, 20, 21, 22 (hours with frequencies above 500,000 under the "Most Orders" category)

"Fewest Orders": 0, 1, 2, 3, 4, 5, 6, 23 (hours with frequencies under 500,000)

```python
result_3 = []

for value in df_ords_prods_merged['order_hour_of_day']:
    if value in range(10, 16):
        result_3.append("Most Orders")
    elif value in (7, 8, 9, 17, 18, 19, 20, 21, 22):
        result_3.append("Average Orders")
    else:
        result_3.append("Fewest Orders")
```

```python
result_3[:20]
```

Creating a new column in the dataframe and labeling it "busiest_period_of_day".
I will then assign the values in the list result_3 to this new column.

```python
df_ords_prods_merged['busiest_period_of_day'] = result_3
```
